```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
path1 = open('data/Table_1_Periodic.txt','r')
path2 = open('data/Table_2_Non_Periodic.txt', 'r')
```

```python
path1_header = path1.readline() 
path1_header 
```

```python
table1 = pd.read_csv('data/Table_1_Periodic.txt')
table1
```

```python
table2 = pd.read_csv('data/Table_2_Non_Periodic.txt')
table2
```

PLOTTING THE GRAPHS!

```python
tem_1 = table1['Teff'] 
rot_1 = table1['Prot'] 
tem_2 = table2['Teff'] 
rot_2 = table2['Prot'] 
plt.figure(figsize=(14.5,10))
plt.scatter(tem_1,rot_1, s=3, c = 'maroon', label = 'Periotic')
plt.scatter(tem_2,rot_2, s=3, c = 'gold',label ='Non_Periotic') 
plt.xlabel('Effective Temperature', fontsize = 25)
plt.ylabel('Rotation Period', fontsize = 25) 
plt.legend()
plt.xlim(6500,3500)
plt.title('Stellar Rotation Period', c = 'navy', fontsize = 30)
```

```python
table1 = table1.dropna()
table2 = table2.dropna() 
len(table1), len(table2)
```

```python
table_1_ar = np.repeat(1,len(table1)) 
print(table_1_ar)
```

```python
table_2_ar = np.repeat(0,len(table2))
print(table_2_ar)
```

```python
classes = np.append(table_1_ar, table_2_ar  )
print(classes)
```

```python
tables = pd.concat([table1, table2], axis=0, sort=True)# sort gives it new indices
tables = tables.drop(columns=['Flag', 'Rper'])
tables
```

```python
tables['class'] = classes
# type(tables)
tables
```

```python

plt.figure(figsize=(14,8))
plt.scatter(tables['Teff'],tables['Prot'],s=.2, c =tables['class'], vmin=0,vmax=1)
plt.colorbar().set_label('Class',fontsize=20)
plt.xlabel('Effective Temperature', fontsize = 25)
plt.ylabel('Rotation Period', fontsize = 25) 
plt.legend()
plt.xlim(6500,3500)
plt.title('Decision Tree Sample of Stellar Rotation Period', c = 'navy', fontsize = 30)
```

```python
tables = tables.drop(columns=['class'])
tables
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = tables 
y = classes
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state = 11)
model = DecisionTreeClassifier()
#fit the model to the training set
fit_model = model.fit(train_X,train_y)
#compare predictions from this model to the test set
predicted = fit_model.predict(test_X)
accuracy_score(test_y, predicted)
```

```python
from sklearn import tree
plt.figure(figsize= (20,15))
tree.plot_tree(model.fit(test_X,test_y)); 
```
