```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
MQ_data = pd.read_csv('data/McQuillan_data.csv')
```

```python
MQ_data
#Naming variables
mass = MQ_data['Mass']
rp = MQ_data['Prot']
temp = MQ_data['Teff']
```

```python
plt.figure(figsize=(15,10))
plt.scatter(temp, rp, c = mass, s = 0.2)
plt.xlim(6500,3000)
plt.colorbar().set_label(label = 'Mass', size = 20)
plt.xlabel("Temperature(k)", size = 30)
plt.ylabel("Rotattion Period", size = 30)
plt.title("McQuillian 2014 data", size = 30)
```
