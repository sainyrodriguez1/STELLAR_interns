```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
from sklearn.tree import DecisionTreeClassifier
import random
from sklearn import tree
```

```python
#generating fake data
#features
temps = []  
spec_type = []
luminosity = []
mass = []
#label
ages = [] 

for i in range(0,100): 
    temps_ran = random.randint(3500, 60000)
    spec_ran = random.randint(0, 7)
    lum_ran = random.randint(1, 100000)
    mass_ran = random.randint(1, 100) 
    ages_ran = random.randint(100000, 13000000) 
    #adding the data to the empty lists
    temps.append(temps_ran)
    spec_type.append(spec_ran)
    luminosity.append(lum_ran)
    mass.append(mass_ran)
    ages.append(ages_ran) 
```

```python
#printing lenghts
print(len(temps),len(spec_type),len(luminosity), len(mass),len(ages))
```

```python
d = {'Tem': temps , 'Spec': spec_type, 'Lum': luminosity, 'Mass': mass}
data_df = pd.DataFrame(data = d)
data_table = Table([temps,spec_type,luminosity,mass],names=('temps','spec_type','luminosity','mass'))
correct_labels = ages
```

```python
data_df, data_table, correct_labels
```

## first tree in forest

```python
#seeds allows the output to be the same, doesn't matter how much you run the cell
random.seed(2)

indices = []
for i in range(0,10):
    indices.append(random.randint(0,100))

data_subset = []
labels_subset = []
for i in indices:
    print(i)
    data_subset.append(data_table[i])
    labels_subset.append(data_table[i])
```

```python
#defining the model
model = DecisionTreeClassifier()
#fitting the model
subset_tree_fit = model.fit(data_subset, labels_subset)
```

```python
#visualizing the tree
plt.figure(figsize=(20,15))
tree.plot_tree(subset_tree_fit);
```
